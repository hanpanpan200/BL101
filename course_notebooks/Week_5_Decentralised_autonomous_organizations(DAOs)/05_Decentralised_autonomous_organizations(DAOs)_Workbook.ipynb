One of the most fundamental decisions that you will have to make is how your DAO will be governed. Like we covered within the coursebook, the voting mechanism to determine voting power heavily influences the kind of DAO you would like to create. Aside from a few other options, Aragon gives you 3 main options.

<center><img src="https://www.dropbox.com/s/wwvgx3ajwg5wfir/type_of_dao.png?raw=1" alt="dao_creation" width="600"/></center>

- **Company**: This DAO utilizes the transferrable token voting power mechanism. Essentially, the more governance tokens you have, the more influential you will become on the DAO's decisions. In this case, since tokens can be bought, influence can be bought as well.
- **Membership**: This DAO utilizes the one-member-one-vote voting mechanism. This is in line with the proof-of-personhood participation model that we discussed. However, since one member could create several wallets, this is not bullet proof.
- **Reputation**: This DAO also utilizes the non-tranferrable token voting power voting mechanism. While the amount of tokens you own, also represent the amount of influence you have, you are not able to buy or sell the token. The token is send or erased from your wallet based on your positive or negative impact on the DAO. This mechanism hols similar elements to the skin-in-the-game model that we discussed.

Go through the rest of the process to launch your DAO. If you need any help you may also inquire the [Aragon Documentation](https://documentation.aragon.org/products/aragon-client/how-to-create-a-dao-using-aragon-client).

#### 3Ô∏è‚É£.2Ô∏è‚É£ Value Model üèÜ  
>In the value model we want to answer a few core questions:  
- **Core Philosophy:** *What's the long term hard problem your business is solven?*  
- **Core Values:** *What key values drive your business idea?*  
- **Value Propoition:** *What set of values each stakeholder gets from your business?*   

#### 3Ô∏è‚É£.3Ô∏è‚É£ Blockchain Model  ‚õì‚öôÔ∏è
>In the Blockchain Model we want to understand how the protocol (which blockchain, what layer, etc.) connects the value model to everything else. 
- **Protocols Rules:** *What rules govern the Blockchain?*  
- **Network Shape:** *What network shape do the protocol rules determine?*  
- **Applications Rules:** *Which applications can be built on top of the protocol?*   

#### 3Ô∏è‚É£.4Ô∏è‚É£ Distribution Model üöö
>Here we want to understand how the various pieces come together to make the underlying Blockchain ecosystem sustainable and keep growing it.
- **Developer Community:** *Is the developer community engaged in developing apps on top of the Blockchain? What does it take to keep it engaged?*  
- **Investor Relations:** *Are the underlying crypto token valuable for potential investors?*   
- **Mining Incentives:** *What incentives exist for miners to keep contributing to the development and growth of the Blockchain network?*  
- **Deal Making:** *What agreements are in place integrating the protocols through institutional & non-institutional channels?*  


#### 3Ô∏è‚É£.5Ô∏è‚É£ Economic Model üìäüìà 
>How does the underlying crypto asset gain value & how do the apps monetize?   

- How is the Blockchain crypto asset gaining value and how the key players part of the ecosystem are monetizing the applications built on top of it? Or very simple: *how do App users make money?*
